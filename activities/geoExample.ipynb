```python
import pandas as pd
import geopandas as geo
import matplotlib.pyplot as plt
from geodatasets import get_path
```

```python
# Create your df
df = pd.DataFrame(
    {
        "City": ['New York', 'Los Angeles', 'Chicago', 'Miami', 'Dallas'],
        "Country": ['United States'] * 5,
        "Latitude": ['40.71', '34.05', '41.88', '25.76', '32.78'],
        "Longitude": ['-74.01', '-118.24', '-87.63', '-80.19', '-96.80'],
    }
)
```

```python
# Convert to a GeoDataFrame
gdf = geo.GeoDataFrame(
    df,
    geometry=geo.points_from_xy(df.Longitude, df.Latitude),
    crs="EPSG:4326",
)
```

```python
world = geo.read_file(get_path("naturalearth.land"))
# Restrict to the region you desire
ax = world.clip([-130, 20, -60, 55]).plot(color="white", edgecolor="black")

#Plot your df
gdf.plot(ax=ax, color="red")

plt.title("Major Cities in the United States")
plt.show()
```
