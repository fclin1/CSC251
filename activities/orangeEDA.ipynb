# Orange EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
orange = pd.read_csv('Orange.csv')
orange.head()
```

```python
orange.tail()
```

```python
orange
```

```python
import seaborn as sns
sns.pairplot(data = orange, x_vars =['circumference'], y_vars = ['age'], hue='Tree', height=5)
```

```python
orange[['circumference', 'age']].corr()
sns.heatmap(orange[['circumference', 'age']].corr(), annot=True, cmap='Blues')
```

```python
r = orange['age'].corr(orange['circumference'])
print(f"Pearson correlation (r) between age and circumference: {r:.4f}")
```

```python
# Regression plot for age vs circumference
sns.lmplot(x='age', y='circumference', data=orange, height=6)
plt.title('Regression Plot of Age vs Circumference')
plt.xlabel('Age')
plt.ylabel('Circumference')
plt.show()
```

```python
# Equation of regression line
import statsmodels.formula.api as smf
result=smf.ols(formula='age ~ circumference', data= orange).fit()
result.summary()
```

```python
# Print the equation of the regression line for the trees in the Orange dataset, where age is the independent variable and circumference is the dependent variable. Include all decimal values given.
intercept = result.params['Intercept']
slope = result.params['circumference']
print(f"Equation of the regression line: circumference = {intercept:.4f} + {slope:.4f} * age")
```
