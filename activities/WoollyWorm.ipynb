
CSC 295 NumPy Activity - Woolly Worm Race

```python
#Import numpy
import numpy as np
#Define constants
TOTALDISTANCE = 42
INTERVAL = 10
TOTALTIME = 90
```

```python
#Randomly generate a 2D array to describe the distance each woolly worm travels in each 10 sec interval (max of 8 in/10 sec)
dist = np.random.randint(80, size = (12,9))
dist = dist / INTERVAL
dist
```

```python
#Compute the cumulative distance (use cumsum method) for each worm at each interval of 10 sec ( use axis = 1 to sum across the columns)
totaldist = np.cumsum(dist, axis = 1)
```

```python
#Import matplotlib and plot the cumulative distance array for each worm by row (transpose using T method)
import matplotlib.pyplot as plt
plt.plot(totaldist.T)
```

```python
#Compute the average speed for each worm over the 90-sec race (total distance/total time)
avgspeed = totaldist[:,-1] / TOTALTIME
avgspeed
```

```python
#Compute the average speed over each 10 sec time interval (ds/dt)
speed10 = dist / INTERVAL
speed10
```

```python
#Which worm won? Remove all doubt by calculating the time to cross the finish line (42 in)
#The time to reach 42 inches = (42 - distance after 80 (?) sec)/average speed in last 10 seconds
lastinterval = (TOTALDISTANCE - totaldist[:,8]) / speed10[:,8]
lastinterval
```

```python
#Confirm max speed (expect .8 in/sec)
speed10max = np.max(speed10[:,:])
speed10max
```

```python
#Confirm min speed (expect 0.0 in/sec)
speed10min = np.min(speed10[:,:])
speed10min
```

```python
#Calculate total time of winner
wtime = (TOTALTIME - INTERVAL) + lastinterval.min()
wtime
```
