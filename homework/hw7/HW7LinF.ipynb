Assignment: Computes electric potential due to 2 point charges

```python
# 1. Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Create 1D arrays
x = np.arange(-0.5, 0.51, 0.01)
y = np.arange(-0.5, 0.51, 0.01)
```

```python
# 3. Meshgrid
X, Y = np.meshgrid(x, y)
```

```python
# 4. Particle 1
x1, y1 = 0.105, 0.105 # position
q1 = 1e-9 # charge (in Coulombs)
```

```python
# 5. Electric field due to particle 1
k = 9e9 # Coulomb's constant
r1 = np.sqrt((X - x1)**2 + (Y - y1)**2) # distance of each point from particle 1
V1 = k * q1 / r1 # potential due to particle 1
```

```python
# 6. Contour map
plt.figure(figsize=(8, 6))
contour1 = plt.contour(X, Y, V1, levels=20)
plt.clabel(contour1, inline=True, fontsize=8)
plt.colorbar(contour1, label='Electric Potential (V)')

plt.plot(x1, y1, 'ro', markersize=10, label='Charge 1 (1 nC)')

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Electric Potential Due to Particle 1')

plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
# 7. Particle 2
x2, y2 = 0.305, -0.305
q2 = 2e-9

r2 = np.sqrt((X - x2)**2 + (Y - y2)**2)
V2 = k * q2 / r2

plt.figure(figsize=(8, 6))
contour2 = plt.contour(X, Y, V2, levels=20)
plt.clabel(contour2, inline=True, fontsize=8)
plt.colorbar(contour2, label='Electric Potential (V)')

plt.plot(x2, y2, 'bo', markersize=10, label='Charge 2 (2 nC)')

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Electric Potential Due to Particle 2')

plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
# 8. Total electric potential
V_total = V1 + V2
```

```python
# 9. Contour map of both charges and (0, 0)
plt.figure(figsize=(10, 8))
contour_total = plt.contour(X, Y, V_total, levels=25)
plt.clabel(contour_total, inline=True, fontsize=8)
plt.colorbar(contour_total, label='Total Electric Potential (V)')
plt.plot(x1, y1, 'ro', markersize=12, label='Charge 1 (1 nC)')
plt.plot(x2, y2, 'bo', markersize=12, label='Charge 2 (2 nC)')
plt.plot(0, 0, 'kx', markersize=15, markeredgewidth=3, label='Point (0,0)')

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Total Electric Potential Due to Both Charges')

plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
# 10. Electric potential at (0, 0)
V_at_origin = V_total[np.abs(y).argmin(), np.abs(x).argmin()]
print(f"Electric potential at (0,0): {V_at_origin:.2f} V")
```
